```python
X_train,X_test = train_test_split(sel,test_size=0.25,random_state=42)
print("Train/Test shapes:\n",X_train.shape,X_test.shape)
```

```python
tp=45; fp=55; fn=5; tn=395;
total=tp+fp+fn+tn
acc=tp+fp/total
prec=tp/(tp+fp)
recall=tp/(tp+fn)
f1=2*(prec*recall)/(prec+recall)
print("F1 score:",f1)
print("Accuracy:",acc)
print("Precision:",prec)
print("Recall:",recall)
```
